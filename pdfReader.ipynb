```python
# THESE ARE THE DEPENDENCIES YOU NEED TO RUN THS CODE. ENSURE ITS INSTALLED.
#!pip install spacy spacy-layout pdfminer.six
#!python -m spacy download en_core_web_sm

import spacy
from spacy_layout import spaCyLayout

nlp = spacy.load("en_core_web_sm")
layout = spaCyLayout(nlp)

#THIS IS MY PDF PATH ON THE INPUT. INI ACTUAITLY, WE WOULD HAVE TO MAKE THIS AN INPUTTED FOR THE CODE
pdf_path = "/content/drive/MyDrive/Colab Notebooks/IntroToAI/Full-Body-MRI-Sample-Report.pdf"
doc = layout(pdf_path)
print(doc.text)


#output it into an actual text file
output_txt_path = "/content/drive/MyDrive/Colab Notebooks/IntroToAI/READ_Full-Body-MRI-Sample-Report.txt"
with open(output_txt_path, "w", encoding="utf-8") as f:
    f.write(doc.text)
```
